```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('tmp/temblores_mexico_50_anios.csv')

df.head()
```

Visualizacion

```python
ubicaciones_count = df['Ubicación'].value_counts()
ubicaciones_count
```

```python
# Crear contexto de figura
plt.figure(figsize=(10,8))

# Configurar seaborn de donde va a tomar la data
sns.barplot(
    x=ubicaciones_count.index,
    y=ubicaciones_count.values,
    palette='viridis',
    legend=False,
    hue=ubicaciones_count.index
)

# Terminar de estilizar nuestro plot
plt.title('Cantidad de temblores por ciudad')
plt.xlabel('Ciudad') # Fontsize
plt.ylabel('Cantidad de temblores')

plt.xticks(rotation=45)
plt.tight_layout()

# Mostrarmos el plot
plt.show()
```
